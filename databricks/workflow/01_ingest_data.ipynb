This Databricks notebook simply ingests an example dataset (titanic data) and stores the raw form in a Delta table.

```python
import pandas as pd
```

```python
#   Ingest from the source URL. Normally a pure Pyspark approach would be better,
#   but since we're pulling from a URL, I'll use pandas and then convert to PySpark
source_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic_df = spark.createDataFrame(pd.read_csv(source_url))

#   Sanity check: inspect the top of the table
display(titanic_df.limit(5))
```

```python
#   Write to Delta table. Ensure the schema exists in the workspace
spark.sql("CREATE DATABASE IF NOT EXISTS bronze")
(
    titanic_df
    .write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("bronze.raw_titanic")
)
```
